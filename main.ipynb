# converting date column from str to date

```python
super_store['Order Date'] = pd.to_datetime(super_store['Order Date'])
super_store['Ship Date'] = pd.to_datetime(super_store['Ship Date'])
```

# 3 new  column for year month week analysis


```python
super_store['Order Year'] = super_store['Order Date'].dt.year
super_store['Order Month'] = super_store['Order Date'].dt.month
super_store['Order day of week'] = super_store['Order Date'].dt.dayofweek

super_store.head()
```

# Monthly Sales Analysis

```python
sale_by_month = super_store.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
sale_by_month
```

```python
fig = px.line(sale_by_month ,
              x = 'Order Month',
              y = 'Sales',
              title= 'Monthly Sales Analysis')

pio.renderers.default = "browser"

fig.show()

```

# Sales by Category

```python
sales_by_category = super_store.groupby('Category')['Sales'].sum().reset_index()

fig = px.pie(sales_by_category,
             names='Category',
             values='Sales',
             hole= 0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition = 'inside' , textinfo = 'percent+label')
fig.update_layout(title='Sales_by_Category' ,title_font = dict(size=24))

pio.renderers.default = 'browser'
fig.show()
```

# Sales and Profit by Segment

```python
sales_profit_by_segment = super_store.groupby('Segment').agg({'Profit':'sum' , 'Sales' : 'sum'}).reset_index()
sales_profit_by_segment
```

```python
color_palette = colors.qualitative.Pastel

fig = go.Figure()

fig.add_trace(go.Bar(x = sales_profit_by_segment['Segment'],
                     y = sales_profit_by_segment['Sales'],
                     name = 'Sales',
                     marker_color = color_palette[0]))
fig.add_trace(go.Bar(x = sales_profit_by_segment['Segment'],
                     y = sales_profit_by_segment['Profit'],
                     name = 'Profit',
                     marker_color = color_palette[1]))

fig.update_layout(title = 'sales profit by segment',
                  xaxis_title = 'Customer Segment' , yaxis_title = 'Amount' , barmode = 'group')

pio.renderers.default = 'browser'

fig.show()
```

# sales by profit ratio

```python
sales_profit_by_segment = super_store.groupby('Segment').agg({'Profit': 'sum' , 'Sales': 'sum'})
sales_profit_by_segment['Sales_to_profit_Ratio'] = sales_profit_by_segment['Profit']/sales_profit_by_segment['Sales']
print(sales_profit_by_segment['Sales_to_profit_Ratio'])
```
